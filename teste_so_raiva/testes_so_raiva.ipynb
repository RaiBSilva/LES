<h3>deslogar_conta_atual()</h3>

```python
def deslogar_conta_atual(browser):
    print("[func]-->deslogar_conta_atual()")
    
    big_timer = Timer(30)
    while big_timer.not_expired:
        timer = Timer(10)
        while timer.not_expired:
            try:
                browser.find_element_by_id('dropdownMenuLink').click()
                break
            except Exception as e:
                continue
        if timer.expired:
            print("Timeout. Drop down da conta logada.")
            continue

        timer.reset()
        while timer.not_expired:
            try:
                options = browser.find_elements_by_tag_name("a")
                for op in options:
                    txt = op.get_attribute("innerText")
                    if txt == "Logout":
                        op.click()
                        return
                else:
                    continue
            except Exception as e:
                continue
        if timer.expired:
            print("Timeout. Não achou o menu logout.")
            continue
    if big_timer.expired:
        raise Exception("Timeout. Não fez logout.")
        
```

<h3>clica_tag_por_txt()</h3>

```python
def clica_tag_por_txt(browser, *, txt:str, tag:str, time=15):
    print("[func]-->clica_tag_por_txt()")
    
    timer = Timer(time)
    while timer.not_expired:
        try:
            options = browser.find_elements_by_tag_name(tag)
            for option in options:
                if txt == option.get_attribute("innerText").strip():
                    option.click()
                    break
            else:
                continue
            break
        except Exception as e:
            continue
    if timer.expired:
        raise Exception(f"Timeout. Não achou o '{txt}' em uma tag <{tag}>.")
```
