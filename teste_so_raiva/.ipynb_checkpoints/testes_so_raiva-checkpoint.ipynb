```python
def test_click_saiba_mais(self):
        browser = ChromeBrowser().opened()
        timer = Timer(15)
        while timer.not_expired:
            try:
                btns = browser.find_elements_by_tag_name("button")
                for btn in btns:
                    texto_btn = btn.get_attribute("innerText")
                    if texto_btn == "SAIBA MAIS":
                        btn.click()
                        break
                else:
                    continue
                break
            except:
                continue
        if timer.expired:
            raise Exception("Timeout. Não achou o btn SAIBA MAIS.")
        time.sleep(4)
        assert browser.title == "Descrição - SóRaiva"
```
