<h3>deslogar_conta_atual()</h3>

```python
def deslogar_conta_atual():
    print("[func]-->deslogar_conta_atual()")
    
    browser = ChromeBrowser().opened()
    timer = Timer(30)
    while timer.not_expired:
        try:
            browser.find_element_by_id('dropdownMenuLink').click()
            break
        except Exception as e:
            continue
    if timer.expired:
        raise Exception("Timeout. Drop down da conta logada.")
        
        
    timer.reset()
    while timer.not_expired:
        try:
            options = browser.find_elements_by_tag_name("a")
            for op in options:
                txt = op.get_attribute("innerText")
                if txt == "Logout":
                    op.click()
                    break
            else:
                continue
            break
        except Exception as e:
            continue
    if timer.expired:
        raise Exception("Timeout. Não achou o menu logout.")
```
